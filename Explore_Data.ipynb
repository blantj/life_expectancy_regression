# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Obtain Data

Read life expectancy dataset into pandas

```python
df = pd.read_csv('excel/scrubbed_dataset.csv')
df.set_index('Country', drop=True, inplace=True)
df.head()
```

# Explore Data

```python
df['Life expectancy at birth, total (years)'].describe()
```

Create histogram of life expectancy distribution

```python
sns.distplot(df['Life expectancy at birth, total (years)'], kde=True, bins=25)
```

Plot relationship between each feature and dependent variable

```python
features = df.columns
for feat in features:
    sns.pairplot(x_vars=feat, y_vars='Life expectancy at birth, total (years)', data=df, kind='reg')
```
